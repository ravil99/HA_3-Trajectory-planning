# 2. Synchroning 6 joints to start and end motion at the same time.

# 2.1 Synchronization

```python
# refine time parameters

t1 = max(t11, t12, t13, t14, t15, t16)
print(f"J1 rise time: {t11}s, J2 rise time: {t12}s, J3 rise time: {t13}s"
      f"J4 rise time: {t14}s, J5 rise time: {t15}s, J6 rise time: {t16}s"
      f"synchronized rise time: {t1}\n")

dwell = max(T11 - t11, T12 - t12, T13 - t13, T14 - t14, T15 - t15, T16 - t16)

print(f"J1 dwell time: {T11-t11}s, J2 dwell time: {T12-t12}s, J3 dwell time: {T13-t13}s"
      f"J4 dwell time: {T14-t14}s, J5 dwell time: {T15-t15}s, J6 dwell time: {T16-t16}s"
      f"synchronized dwell time: {dwell}\n")

T = dwell+t1
tf = T + t1
print("Synchronized trajectory time = {} + {} + {} = {}s\n".format(t1, dwell, t1, tf))

t_params = [t0, t1, T, tf]
print(f't_params: {t_params}')

# recalculate velocity parameters
dq1_m = (j1[1]-j1[0])/T
dq2_m = (j2[1]-j2[0])/T
dq3_m = (j3[1]-j3[0])/T
dq4_m = (j4[1]-j4[0])/T
dq5_m = (j5[1]-j5[0])/T
dq6_m = (j6[1]-j6[0])/T

# recalculate acceleration parameters
ddq1_m = dq1_m/t1
ddq2_m = dq2_m/t1
ddq3_m = dq3_m/t1
ddq4_m = dq4_m/t1
ddq5_m = dq5_m/t1
ddq6_m = dq6_m/t1


print("Joint 1 velocity modified from {} to {} and acceleration from {} to {}".format(
    j1[2], dq1_m, j1[3], ddq1_m))
print("Joint 2 velocity modified from {} to {} and acceleration from {} to {}".format(
    j2[2], dq2_m, j2[3], ddq2_m))
print("Joint 3 velocity modified from {} to {} and acceleration from {} to {}".format(
    j3[2], dq3_m, j3[3], ddq3_m))
print("Joint 4 velocity modified from {} to {} and acceleration from {} to {}".format(
    j4[2], dq4_m, j4[3], ddq4_m))
print("Joint 5 velocity modified from {} to {} and acceleration from {} to {}".format(
    j5[2], dq5_m, j5[3], ddq5_m))
print("Joint 6 velocity modified from {} to {} and acceleration from {} to {}".format(
    j6[2], dq6_m, j6[3], ddq6_m))

# update parameters
j1[2:] = [dq1_m, ddq1_m]
j2[2:] = [dq2_m, ddq2_m]
j3[2:] = [dq3_m, ddq3_m]
j4[2:] = [dq4_m, ddq4_m]
j5[2:] = [dq5_m, ddq5_m]
j6[2:] = [dq6_m, ddq6_m]
```

# 3. Numerical control

```python
delta_t = 0.002

if t1 % delta_t != 0:
    num_t1 = (t1 // delta_t) * delta_t + delta_t

if T % delta_t != 0:
    T = (T // delta_t) * delta_t + delta_t

if tf % delta_t != 0:
    tf = (tf // delta_t) * delta_t + delta_t

t_params = [t0, t1, T, tf]
print("Synchronized trajectory time = {} + {} + {} = {}s\n".format(t1, T - t1, t1, tf))

# recalculate velocity parameters
dq1_m = (j1[1]-j1[0])/T
dq2_m = (j2[1]-j2[0])/T
dq3_m = (j3[1]-j3[0])/T
dq4_m = (j4[1]-j4[0])/T
dq5_m = (j5[1]-j5[0])/T
dq6_m = (j6[1]-j6[0])/T

# recalculate acceleration parameters
ddq1_m = dq1_m/t1
ddq2_m = dq2_m/t1
ddq3_m = dq3_m/t1
ddq4_m = dq4_m/t1
ddq5_m = dq5_m/t1
ddq6_m = dq6_m/t1

# update parameters
j1[2:] = [dq1_m, ddq1_m]
j2[2:] = [dq2_m, ddq2_m]
j3[2:] = [dq3_m, ddq3_m]
j4[2:] = [dq4_m, ddq4_m]
j5[2:] = [dq5_m, ddq5_m]
j6[2:] = [dq6_m, ddq6_m]

print('q_0, q_f, v_max, a_max')
print(f"{j1}\n {j2}\n {j3}\n {j4}\n {j5}\n {j6}")
```

```python
t0 = 0
t1 = 14.608
t2 = 2*t1

coef = np.zeros((4,4), dtype = np.float64)
coef[0,:] = [ t0 ** 3, t0 ** 2, t0 , 1 ]
coef[1,:] = [ t1 ** 3, t1 ** 2, t1 , 1 ]
coef[2,:] = [ 3 * t0 ** 2, 2* t0, 1 , 0 ]
coef[3,:] = [ 3 * t1 ** 2, 2* t1, 1 , 0  ]

coef_2 = np.zeros((4,4), dtype = np.float64)
coef_2[0,:] = [ t1 ** 3, t1 ** 2, t1 , 1 ]
coef_2[1,:] = [ t2 ** 3, t2 ** 2, t2 , 1  ]
coef_2[2,:] = [ 3 * t1 ** 2, 2* t1, 1 , 0]
coef_2[3,:] = [ 3 * t2 ** 2, 2* t2, 1 , 0 ]

# q0, qf, v0, vf, a0, af
q_0 = 0
q_1 = 90
q_2 = 120

v_0 = 0
v_1 = 2
v_2 = 0


first_const = [q_0, q_1, v_0, v_1]
second_const = [q_1, q_2, v_1, v_2]

trj_first = np.matmul(np.linalg.inv(coef), first_const)

trj_second = np.matmul(np.linalg.inv(coef_2), second_const)

print(type(trj_first))
print(trj_first)

# From point 1 to point 2
T1,q1,v1,a1 = f.cubic_polynom(trj_first, t0, t1)

T1 = list(T1)

print(f'T1: {len(T1)}, {type(T1)}, {T1[0]}, {T1[-1]}  \n')
print(f'q1: {len(q1)}, {type(q1)}, {q1[0]}, {q1[-1]}  \n')
print(f'v1: {len(v1)}, {type(v1)}, {v1[0]}, {v1[-1]}  \n')
print(f'a1: {len(a1)}, {type(a1)}, {a1[0]}, {a1[-1]}  \n')

# From point 2 to point 3
T2,q2,v2,a2 = f.cubic_polynom(trj_second, t1, t2)

T2 = list(T2)

print(f'T2: {len(T2)}, {type(T2)}, {T2[0]}, {T2[-1]}  \n')
print(f'q2: {len(q2)}, {type(q2)}, {q2[0]}, {q2[-1]}  \n')
print(f'v2: {len(v2)}, {type(v2)}, {v2[0]}, {v2[-1]}  \n')
print(f'a2: {len(a2)}, {type(a2)}, {a2[0]}, {a2[-1]}  \n')
```
